```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visualization style
sns.set(style="whitegrid")
%matplotlib inline

# Load the dataset (you can upload your file or use this sample data)
data = """
PassengerId,Survived,Pclass,Name,Sex,Age,SibSp,Parch,Ticket,Fare,Cabin,Embarked
1,1,3,"Allen, Mr. William",male,35,0,0,373450,8.05,,S
2,0,1,"Bonnell, Miss. Elizabeth",female,58,0,0,113781,26.55,C103,C
3,1,3,"Saundercock, Mr. William H.",male,20,0,0,A/5. 2151,7.25,,S
4,1,1,"Andersson, Mr. Anders Johan",male,39,1,5,347082,31.28,,S
5,0,3,"Vestrom, Miss. Hulda Amanda Adolfina",female,14,0,0,350406,7.75,,S
6,0,3,"Hewlett, Mrs. (Mary D Kingcome)",female,55,0,0,248706,16,,S
7,1,2,"Rice, Master. Eugene",male,2,4,1,382652,29.12,,Q
8,0,2,"Williams, Mr. Charles Eugene",male,30,0,0,244373,13,,S
9,1,1,"Futrelle, Mrs. Jacques Heath (Lily May Peel)",female,36,1,0,113803,53.1,C123,C
10,0,3,"McCarthy, Mr. Timothy J.",male,54,0,0,17463,51.86,E46,S
"""

# Create DataFrame
from io import StringIO
df = pd.read_csv(StringIO(data))

# Display basic information
print("=== Dataset Overview ===")
print(f"Number of passengers: {len(df)}")
print("\n=== First 5 Rows ===")
display(df.head())

print("\n=== Dataset Information ===")
display(df.info())

print("\n=== Descriptive Statistics ===")
display(df.describe(include='all'))

# Visualization 1: Survival by Class
plt.figure(figsize=(10, 5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()

# Visualization 2: Survival by Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()

# Visualization 3: Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Visualization 4: Fare Distribution by Class
plt.figure(figsize=(10, 5))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='Set3')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

# Visualization 5: Age vs Fare colored by Survival
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', style='Pclass', data=df, s=100)
plt.title('Age vs Fare (Colored by Survival, Styled by Class)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Additional Analysis
print("\n=== Survival Rate by Class ===")
display(pd.crosstab(df['Pclass'], df['Survived'], normalize='index') * 100)

print("\n=== Survival Rate by Gender ===")
display(pd.crosstab(df['Sex'], df['Survived'], normalize='index') * 100)

print("\n=== Average Age by Survival ===")
display(df.groupby('Survived')['Age'].mean())

print("\n=== Average Fare by Survival ===")
display(df.groupby('Survived')['Fare'].mean())

# Family size analysis
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print("\n=== Family Size Distribution ===")
display(df['FamilySize'].value_counts())

plt.figure(figsize=(10, 5))
sns.countplot(x='FamilySize', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```
